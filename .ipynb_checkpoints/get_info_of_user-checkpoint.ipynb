```python
from bs4 import BeautifulSoup
import requests
import re
import json

r = requests.get('https://www.instagram.com/5479.sm/')
soup = BeautifulSoup(r.text)
scripts = soup.find_all('script')


for script in scripts:
#     print(script.text.find("window._sharedData"))
#     print(script)
#     print("\n\n")
    if (script.text.find("window._sharedData =") >= 0):
#         print(script)
        break


# print(scripts)

variable_start = script.text.find('{')
variable_end = script.text.rfind('}')+1

userPage = json.loads(script.text[variable_start:variable_end])
userInfo = userPage["entry_data"]["ProfilePage"][0]["graphql"]["user"]
print(json.dumps(userInfo, indent=4, sort_keys=True))
print("Username is: " + userInfo["username"])
print("Biography is: " + userInfo["biography"])

# print(stock["entry_data"]["ProfilePage"][0]["graphql"]["user"])
# print(json.dumps(stock, indent=4, sort_keys=True))
```
